# CONSTRUCTION OF SVI EQUITY VOL SURFACE

In this notebook I show how you can calibrate to the FX Vol Surface to ATM, 25D MS and 25D RR at one expiry date and analyse different volatility interpolation methods.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from financepy.market.curves.FinDiscountCurveFlat import FinDiscountCurveFlat
from financepy.market.volatility.FinEquityVolSurface import FinEquityVolSurface
from financepy.finutils.FinDate import FinDate
from financepy.models.FinModelVolatilityFns import FinVolFunctionTypes
```

# Load the Volatility Market Quotes

```python
valueDate = FinDate(11, 1, 2021)
```

```python
stockPrice = 3800.0 # Check
```

```python
expiryDates = [FinDate(11, 2, 2021), FinDate(11, 3, 2021), FinDate(11, 4, 2021), 
               FinDate(11, 7, 2021), FinDate(11,10, 2021), FinDate(11, 1, 2022), 
               FinDate(11, 1, 2023)]
```

```python
strikes = np.array([3037, 3418, 3608, 3703, 3798, 3893, 3988, 4178, 4557], dtype=float)
```

Now we pass in the volatility grid

```python
volSurface = [[42.94, 31.30, 25.88, 22.94, 19.72, 16.90, 15.31, 17.54, 25.67],
             [37.01, 28.25, 24.19, 21.93, 19.57, 17.45, 15.89, 15.34, 21.15],
             [34.68, 27.38, 23.82, 21.85, 19.83, 17.98, 16.52, 15.31, 18.94],
             [31.41, 26.25, 23.51, 22.05, 20.61, 19.25, 18.03, 16.01, 15.90],
             [29.91, 25.58, 23.21, 22.01, 20.83, 19.70, 18.62, 16.63, 14.94],
             [29.26, 25.24, 23.03, 21.91, 20.81, 19.73, 18.69, 16.76, 14.63],
             [27.59, 24.33, 22.72, 21.93, 21.17, 20.43, 19.71, 18.36, 16.26]]
```

```python
volSurface = np.array(volSurface)
volSurface = volSurface / 100.0        
```

```python
rfrRate = 0.020  # USD
discountCurve = FinDiscountCurveFlat(valueDate, rfrRate)

divRate = 0.010  # USD
dividendCurve = FinDiscountCurveFlat(valueDate, divRate)
```

# Calibrate Volatility Surface 

```python
volFunctionType = FinVolFunctionTypes.SVI
```

```python
equitySurface = FinEquityVolSurface(valueDate,
                                    stockPrice,
                                    discountCurve,
                                    dividendCurve,
                                    expiryDates,
                                    strikes,
                                    volSurface,
                                    volFunctionType)
```

```python
tol = 1e-4
equitySurface.checkCalibration(False, tol)
```

## Volatility Curves

```python
equitySurface.plotVolCurves()
```

The fit looks so good, it's almost as if everyone is using the SVI parametrisation ;-)
